```python
#자료를 불러 데이터 프레임으로 전환
import pandas as pd

News1 = pd.read_excel("News(1030-1105).xls")
News1.head()
```

```python
#새로운 데이터로 옮기기
News_data = News1[['일자', '언론사', '제목', '키워드']]
News_data.head()
```

```python
#형태소 분석기를 이용하여 분류2
from konlpy.tag import Okt
News_keyword = News_data['키워드']

words2 = []

t = Okt()

for i in News_keyword:
    token_nouns = t.nouns(i)
    #print('token:', token_nouns)
    for temp_nouns in token_nouns:
        if len(temp_nouns) > 1:
            words2.append(temp_nouns)
        
print(words2[:100])
```

```python
#단어: 갯수 형태로 설정
words_cnt = {}
for i in words2:
    words_cnt[i] = words_cnt.get(i, 0) +1

print(words_cnt)
```
